*Часовий ряд* – це набір спостережень $X_t$, кожне з яких записане в
конкретний час $t$. Часовий ряд із дискретним часом – це коли спостереження
здійснюється через фіксовані проміжки часу.  Неперервні часові ряди отримують, коли
спостереження реєструються безперервно протягом деякого інтервалу часу.
— Броквелл і Девіс (2016)

```python
import quandl
import matplotlib.pyplot as plt

from creds import nasdaq_key
oil = quandl.get("NSE/OIL", authtoken=nasdaq_key,#insert you api token
                 start_date="1980-01-01",
                 end_date="2023-01-01")

plt.figure(figsize=(10, 6))
plt.plot(oil.Close)
plt.ylabel('$')
plt.xlabel('Date')
plt.show()
```

## Компоненти часового ряду

Часовий ряд має чотири компоненти, а саме 

- тренд

- сезонність

- циклічність

- шум 

```python
import yfinance as yf
import datetime

ticker = "^GSPC"
start = datetime.datetime(2001, 9,11)
end = datetime.datetime(2023, 9,1)

SP = yf.download(ticker, start = start, end = end, interval = '1mo').Close
```

## Функція автокореляції

```python
from statsmodels.tsa.seasonal import seasonal_decompose
seasonal_decompose(SP, period = 12).plot()
plt.show()
```
